#### Check character presence, index positions, and count

```python

text = input("Enter a string: ")

# Input character to search
ch = input("Enter a character to check: ")

# Check presence
if ch in text:
    print(f"Character '{ch}' is present in the string.")
else:
    print(f"Character '{ch}' is NOT present in the string.")

# Find index positions
positions = []
for i in range(len(text)):
    if text[i] == ch:
        positions.append(i)

# Display positions
if positions:
    print("Index positions:", positions)
else:
    print("No index positions found.")

# Count occurrences
count = 0
for c in text:
    if c == ch:
        count += 1

print("Total count:", count)
```

#### Reverse a list without using reverse()

```python
lst = input("Enter list elements separated by space: ").split()
rev_list = []

for i in range(len(lst)-1, -1, -1):
    rev_list.append(lst[i])

print("Reversed List:", rev_list)
```

#### Find largest number without using max()

```python
numbers = [10, 45, 3, 89, 22]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

#### Rotate list elements

```python
lst = input("Enter list elements separated by space: ").split()

last = lst[-1]
for i in range(len(lst)-1, 0, -1):
    lst[i] = lst[i-1]

lst[0] = last
print("Rotated List:", lst)
```

## Dictionary 

#### Check whether a given key exists in a dictionary

```python
my_dict = {'a': 10, 'b': 20, 'c': 30}
key = 'b'
if key in my_dict:
    print("Key exists in the dictionary")
else:
    print("Key does not exist in the dictionary")
```

#### Sum all the items (values) in a dictionary

```python
my_dict = {'a': 15, 'b': 25, 'c': 30}
total = sum(my_dict.values())
print("Sum of values:", total)
```

#### Multiply all the items (values) in a dictionary

```python
my_dict = {'a': 6, 'b': 3, 'c': 4}
result = 1
for value in my_dict.values():
    result *= value
print("Multiplication of values:", result)
```

#### Sort a dictionary by key

```python
my_dict = {'b': 20, 'a': 10, 'c': 30}
sorted_dict = dict(sorted(my_dict.items()))
print("Sorted dictionary by key:", sorted_dict)
```

#### Get the maximum and minimum value in a dictionary

```python
my_dict = {'a': 10, 'b': 50, 'c': 5}
max_value = max(my_dict.values())
min_value = min(my_dict.values())
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```
